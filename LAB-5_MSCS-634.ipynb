                                            Step 1: Data Preparation and Exploration

```python
# Loading and exploring the data sets

from sklearn.datasets import load_wine
import pandas as pd
import numpy as np

# Load the wine dataset
wine = load_wine()
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data['target'] = wine.target

# Explore the dataset
print(data.head())
print(data.info())
print(data.describe())
```

```python
#Standardising the features which can be used for clustering
from sklearn.preprocessing import StandardScaler

# Drop target for clustering
X = data.drop('target', axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

                                                  Step 2: Hierarchical Clustering

```python
#Applying Agglomerative Clustering on the loaded data sets

from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns

# Trying different cluster values
for n_clusters in [2, 3, 4]:
    model = AgglomerativeClustering(n_clusters=n_clusters)
    labels = model.fit_predict(X_scaled)

    # Visualizing the diffrent Plots for different cluser values
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=labels, palette='viridis')
    plt.title(f'Hierarchical Clustering (n_clusters={n_clusters})')
    plt.xlabel(wine.feature_names[0])
    plt.ylabel(wine.feature_names[1])
    plt.show()
```

```python
#Generating and interpreting a dendrogram to better understand the hierarchical structure.
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(X_scaled, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

                                                      Step 3: DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Try multiple combinations of eps and min_samples
eps_values = [0.3, 0.5, 0.7, 1.0]
min_samples_values = [3, 5, 10]

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        db_labels = db.fit_predict(X_scaled)

        unique_labels = set(db_labels)
        n_clusters = len(unique_labels - {-1})
        n_noise = list(db_labels).count(-1)

        print(f"\n=== DBSCAN (eps={eps}, min_samples={min_samples}) ===")
        print("Cluster label counts:", Counter(db_labels))
        print(f"Number of clusters (excluding noise): {n_clusters}")
        print(f"Number of noise points: {n_noise}")

        # Evaluate clustering
        if n_clusters >= 2:
            filtered_X = X_scaled[db_labels != -1]
            filtered_labels = db_labels[db_labels != -1]
            sil_score = silhouette_score(filtered_X, filtered_labels)
            print("Silhouette Score:", sil_score)
        else:
            print("Silhouette Score: Cannot compute (need at least 2 clusters)")

        print("Homogeneity Score:", homogeneity_score(data['target'], db_labels))
        print("Completeness Score:", completeness_score(data['target'], db_labels))

        # Visualize clusters and highlight noise
        plt.figure(figsize=(6, 4))
        palette = sns.color_palette('Set1', len(set(db_labels)) - (1 if -1 in db_labels else 0))
        sns.scatterplot(
            x=X_scaled[:, 0], y=X_scaled[:, 1], hue=db_labels,
            palette=palette + [(0.6, 0.6, 0.6)] if -1 in db_labels else palette,
            legend='full'
        )
        plt.title(f'DBSCAN Clustering (eps={eps}, min_samples={min_samples})')
        plt.xlabel(wine.feature_names[0])
        plt.ylabel(wine.feature_names[1])
        plt.legend(title='Cluster', loc='upper right', bbox_to_anchor=(1.2, 1))
        plt.show()
```

```python
# As the silhouette score was not getting calcualted even after a lof of parameter tunes
# I have reduded the dinemsionalty to 2D this will allow DBSCAN to reduce lot of noises in high dimensions

from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Dimensionality reduction to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 2: Try DBSCAN on PCA-reduced data
db = DBSCAN(eps=0.5, min_samples=5)
db_labels = db.fit_predict(X_pca)

# Step 3: Evaluation
unique_labels = set(db_labels)
n_clusters = len(unique_labels - {-1})
n_noise = list(db_labels).count(-1)

print(f"\n=== DBSCAN (on PCA) eps=0.5, min_samples=5 ===")
print("Cluster label counts:", Counter(db_labels))
print(f"Number of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {n_noise}")

if n_clusters >= 2:
    filtered_X = X_pca[db_labels != -1]
    filtered_labels = db_labels[db_labels != -1]
    print("Silhouette Score:", silhouette_score(filtered_X, filtered_labels))
else:
    print("Silhouette Score: Cannot compute (need at least 2 clusters)")

print("Homogeneity Score:", homogeneity_score(data['target'], db_labels))
print("Completeness Score:", completeness_score(data['target'], db_labels))

# Step 4: Visualize clusters
plt.figure(figsize=(6, 4))
palette = sns.color_palette('Set1', len(set(db_labels)) - (1 if -1 in db_labels else 0))
sns.scatterplot(
    x=X_pca[:, 0], y=X_pca[:, 1], hue=db_labels,
    palette=palette + [(0.6, 0.6, 0.6)] if -1 in db_labels else palette,
    legend='full'
)
plt.title("DBSCAN on PCA-Reduced Data")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title='Cluster', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

*Hierarchical Clustering Observations*


** Strengths:**

This does not require parameter tuning other than number of clusters.

Provides a dendrogram, which gives insight into data structure.

Works well when clusters are globular or convex and roughly equal in size.



** Weaknesses:**

This requires us to pre-specify number of clusters (k).

This method is not robust to outliers — all data points are assigned to some cluster.

Doesn’t scale well with very large datasets.

In my case I have tried different values for  n_clusters = 2, 3, 4. Visually, 3 gave the best separation, aligning with the actual wine class labels (target).

No noise was detected which is as expected — hierarchical doesn’t filter out outliers.

Moderate performance on homogeneity and completeness which meaning clusters were mixed but somewhat aligned with actual labels.

*DBSCAN Observations *

** Strengths:**

DBSCAN not require number of clusters to be set — automatically detects them.

This also identifies outliers as noise label = -1 which can be useful in real-world noisy data.

Works well with non-globular, irregularly shaped clusters.

Better silhouette score after PCA due to tighter, separated clusters.

** Weaknesses: **
DBSCAN is highly sensitive to eps and min_samples — bad values result in:

All points as noise

Only one cluster

Performs poorly in high-dimensional data → needs PCA or feature selection.

Evaluation metrics like homogeneity may suffer because it doesn’t attempt to match true class labels.

**In my Case:**
In full-dimensional space, DBSCAN failed completely — all points labeled as noise.

After reducing to 2D with PCA, DBSCAN formed meaningful clusters and allowed silhouette computation.

Some points were marked as noise — which show cases DBSCAN’s outlier detection capability.

Lower homogeneity but higher completeness, which indicaties it grouped related data together, but not always uniquely aligned with true labels.

**Conclusion:**

Hierarchical clustering is great when we want a quick overview of the data and the number of clusters is roughly known. It’s simple and interpretable, but not robust to noise.

DBSCAN is ideal when clusters are arbitrary-shaped, and especially when we want to identify outliers. But it requires careful tuning of parameters and works better after dimensionality reduction like PCA.
